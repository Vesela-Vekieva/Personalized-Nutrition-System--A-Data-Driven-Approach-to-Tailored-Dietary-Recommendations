##### 2. Proportional imputation

The idea is to fill missing values in a way that the overall distribution of the categories remains the same. Instead of simply imputing with the mode (the most frequent category), this method ensures that the proportions of the categories remain consistent before and after imputation.

For the columns `diet_type, daily_coffee_consumption, food_intolerances, weekly_exercise_frequency and sleep_quality` I will distribute NaNs equally across other unique categories for the specified columns in the dataset. The function will be applied to all relevant columns to ensure the distribution remains consistent.

````python
# List of columns to get value counts for
columns = ['diet_type', 'daily_coffee_consumption', 'food_intolerances', 
           'weekly_exercise_frequency', 'sleep_quality']

# Loop through the columns and print value counts
for col in columns:
    print(dietry_habits_synthetic_cleaned[col].value_counts(dropna = False))
    print("\n")
````

````python
def distribute_nans_equally(dietry_habits_synthetic_cleaned, column_name):
    # Get the rows with NaN values
    nan_indices = dietry_habits_synthetic_cleaned[dietry_habits_synthetic_cleaned[column_name].isna()].index
    
    # Get the unique values and their counts (excluding NaN)
    unique_values = dietry_habits_synthetic_cleaned[column_name].dropna().unique()
    
    # Calculate the proportion for each unique value
    value_counts = dietry_habits_synthetic_cleaned[column_name].value_counts(normalize = True)
    
    # Determine the number of NaNs to assign to each category
    n_nans = len(nan_indices)
    counts_to_assign = (value_counts * n_nans).round().astype(int)
    
    # Adjust if there's any rounding discrepancy
    discrepancy = n_nans - counts_to_assign.sum()
    if discrepancy != 0:
        counts_to_assign.iloc[0] += discrepancy
    
    # Create a list of values to assign to NaNs
    replacement_values = np.hstack([
        np.repeat(value, count) for value, count in counts_to_assign.items()
    ])
    
    # Shuffle the replacement values to randomize the assignment
    np.random.shuffle(replacement_values)
    
    # Assign the shuffled values to the NaN indices
    dietry_habits_synthetic_cleaned.loc[nan_indices, column_name] = replacement_values

# Columns to apply the function to
columns_to_impute = ['diet_type', 'daily_coffee_consumption', 'food_intolerances', 'weekly_exercise_frequency', 'sleep_quality']

# Apply the function to each column
for column in columns_to_impute:
    distribute_nans_equally(dietry_habits_synthetic_cleaned, column)
````

The code provides a method for handling missing values by distributing them based on the distribution of existing values. This approach ensures that missing values are filled in a manner consistent with the observed data, helping to maintain the overall data distribution and potentially improve the quality of any subsequent analyses or modeling.

##### 4. Convert `daily_coffee_consumption` and `weekly_exercise_frequency` into integer for better analysis

Converting ranges into a single numerical value simplifies the analysis. For example, it makes it easier to calculate means, medians, or correlations with other numerical variables like BMI or weight.
Also, this approach ensures compatibility with models - many statistical and machine learning models require numerical inputs. Converting to integers allows these variables to be used directly in such models.

````python
# The function will handle both ranges and single values:
def convert_range_to_avg(value):
    if '-' in value:
        low, high = map(int, value.split('-'))
        return int((low + high) / 2)
    else:
        return int(value)

dietry_habits_synthetic_cleaned['daily_coffee_consumption'] = (
    dietry_habits_synthetic_cleaned['daily_coffee_consumption'].apply(convert_range_to_avg))

dietry_habits_synthetic_cleaned['weekly_exercise_frequency'] = (
    dietry_habits_synthetic_cleaned['weekly_exercise_frequency'].apply(convert_range_to_avg))
````

This function will take a string value, check if it's a range, and return the average of the two numbers. If it's a single value, it will convert it directly to an integer.
The function uses `int()` to convert the calculated average to an integer. In Python, `int()` truncates the decimal part (rounds down), so the average of 1.5 becomes 1.

__Correlation Analysis__

A correlation matrix is a table showing correlation coefficients between variables. Each cell in the table shows the correlation between two variables. The value is between -1 and 1. If two variables have a high correlation (close to 1 or -1), it implies that as one variable changes, the other variable tends to change in a specific direction.

Correlation Coefficient Interpretation:
    
0.0 to 0.1: Negligible correlation  
0.1 to 0.3: Weak correlation  
0.3 to 0.5: Moderate correlation  
0.5 to 0.7: Strong correlation  
0.7 to 0.9: Very strong correlation  
0.9 to 1.0: Extremely strong correlation 

Values close to 1 or -1 indicate a strong positive or negative correlation, respectively, while values near 0 indicate no linear correlation.

#### Introduction to macronutrients:

Macronutrients (macros) are the nutrients our body needs in large quantities, including protein, carbohydrates and fat. On the other hand, micronutrients consist of different vitamins and minerals and are also essential but consumed in smaller quantities. 

Counting and keeping track of macros can help you reach various health goals.
The United States Department of Agriculture (USDA) has provided a rough guide as to how much of each macronutrient our body requires:

- Carbohydrates: 45-65% of your daily calories

Carbohydrates are a macronutrient that our bodies turn into glucose to use as energy to function. They make up the most considerable portion of one’s calorie intake as they are the most important energy source. 

- Protein: 10-35% of your daily calories

Protein is an essential macronutrient that is highly researched, particularly in sports nutrition and exercise. It is found throughout the body in our muscles, bone, skin and hair and is made up of amino acids, often referred to as the building blocks of protein. There are 20 amino acids, of which nine are essential, meaning our bodies cannot synthesise them so we must get them from the foods in our diet.  

- Fats: 20-35% of your daily calories

While fats are often misunderstood to be bad for us, dietary fat is an essential part of our diet. Dietary fat has two main functions: providing your body with energy and supporting cell function.

This categorization simplifies the analysis and facilitates filtering by food type.

During this process, I also noticed several food categories that are not relevant to my analysis, such as 'babyfood,' 'infant formula,' and various fruits and vegetables that are already covered in a different dataset. To address this, I assigned the value 'none' to these rows in the type column and subsequently removed them from the dataset.

````python
def standardize_type(name):
    # Mapping keywords to their corresponding standardized categories
    mapping = {
        'veggie ': 'vegetarian',
        'coconut milk': 'plant-based drink',
        'soymilk': 'plant-based drink',
        'butter': 'oils and solid fats',
        'oil': 'oils and solid fats',
        'fat': 'oils and solid fats',
        'margarine': 'oils and solid fats',
        'cheese': 'dairy',
        'cream': 'dairy',
        'milk': 'dairy',
        'yogurt': 'dairy',
        'egg': 'egg',
        'turkey': 'meat',
        'poultry': 'meat',
        'chicken': 'meat',
        'chick': 'meat',
        'emu': 'meat',
        'ostrich': 'meat',
        'wurst': 'meat',
        'sausage': 'meat',
        'beef': 'meat',
        'pork': 'meat',
        'lamb': 'meat',
        'veal': 'meat',
        'salmon': 'fish',
        'sea bass': 'fish',
        'shark': 'fish',
        'tuna': 'fish',
        'whitefish': 'fish',
        'crab': 'shellfish',
        'lobster': 'shellfish',
        'shrimp': 'shellfish',
        'clam': 'shellfish',
        'mussle': 'shellfish',
        'oyster': 'shellfish',
        'octopus': 'shellfish',
        'scallop': 'shellfish',
        'squid': 'shellfish',
        'wheat ': 'grains',
        'rice': 'grains',
        'oat': 'grains',
        'corn': 'grains', 
        'cereal': 'grains',
        'grain': 'grains',
        'bagel': 'grains', 
        'macaroni': 'grains',
        'noodles': 'grains',
        'spaghetty': 'grains',
        'tortilla': 'grains',
        'bev': 'beverages',
        'juc': 'beverages',
        'juice': 'beverages',
        'drink': 'beverages',
        'lemonade': 'beverages',
        'tea': 'beverages',
        'coffee': 'beverages',
        'water': 'beverages',
        'drk': 'beverages',
        'cake': 'added sugars',
        'cookies': 'added sugars',
        'pastry': 'added sugars',
        'doughnut': 'added sugars',
        'croissant': 'added sugars',
        'eclair': 'added sugars',
        'muffin': 'added sugars',
        'ice cream': 'added sugars',
        'pancakes': 'added sugars',
        'pie': 'added sugars',
        'waffles': 'added sugars',
        'pudding': 'added sugars'
    }
    
    # Iterate over the mapping and return the standardized type if found
    for key, value in mapping.items():
        if key in name:
            return value
    
    # Return None if no match is found
    return None

# Apply the function to the 'type' column
food_macronutrients['type'] = food_macronutrients['type'].apply(standardize_type)

# Remove rows where 'type' is None
food_macronutrients = food_macronutrients[food_macronutrients['type'].notna()]
````

__Standardization Function__

The function ```standardize_type()``` maps specific keywords found in the type column to broader, standardized food categories.
A dictionary named mapping is used to define these mappings, associating each keyword (e.g., 'butter', 'cheese', 'beef') with a corresponding category (e.g., 'oils and solid fats', 'dairy', 'meat').
The function iterates over each keyword in the mapping. If a keyword is found within the string passed to the function (which is a food type name), it returns the corresponding standardized category.
If no match is found, the function returns None.

The function is applied to the type column in the ```food_macronutrients```. This replaces the original values in the type column with their corresponding standardized categories.

After standardizing the type column, any rows where the type is None (i.e., rows that did not match any keyword in the mapping) are removed from the DataFrame. This step ensures that the dataset only contains relevant food categories.

#### Standardizing the Names
This code processes a column ```name``` in the ```food_macronutrients_merged_data``` DataFrame to extract the main product name and move any additional details to a new description column. Here's a step-by-step explanation

````python
# Function to extract the main name and description
def extract_name_and_description(row):
     # List of separators and keywords for splitting the text
    separators = [',', ';', ':', '-', 'nutrition', 'facts', 'with', 'or', 'and']
    name = row
    description = None
    
    for sep in separators:
        if sep in name:
            parts = name.split(sep, 1)
            name = parts[0].strip()
            description = parts[1].strip() if len(parts) > 1 else None
            break
    
    return pd.Series([name, description])

# Apply the function to the 'name' column and create new columns 'name' and 'description'
food_macronutrients_merged_data[['name', 'description']] = food_macronutrients_merged_data['name'].apply(extract_name_and_description)

# Fill missing descriptions with the main name if the description is None
food_macronutrients_merged_data['description'].fillna(food_macronutrients_merged_data['name'], inplace=True)

print(food_macronutrients_merged_data)
````

### Saving Data with Pickle for future work

#### What is Pickle?

Pickle is a Python-specific library that allows you to serialize (convert) Python objects into a byte stream, which can then be saved to a file, and later deserialized (loaded) back into Python objects. This is especially useful when you are working with complex objects like pandas DataFrames, which may contain various data types, indexes, and other metadata that are difficult to represent in plain text files like CSVs.

#### Why Use Pickle for Saving and Loading Datasets?
1. Efficiency: Pickle is very efficient for saving and loading large or complex Python objects. Unlike CSV or other text-based formats, pickle preserves the data structure and is more compact for storing Python objects.
2. Convenience: When you load the file back, you don’t have to reprocess or reformat the data. It’s restored exactly as it was when it was saved.
3. Compatibility with Python: Pickle is native to Python, so it works seamlessly with Python-based tools like pandas, which makes it an ideal choice for saving DataFrame objects.

````python
# set the directory
directory_path = './cleaned datasets for future work/'

# Save the diet dataset as a pickle file
with open(directory_path + 'dietry_habits_cleaned.pkl', 'wb') as f:
    pickle.dump(dietry_habits_synthetic_cleaned, f)

# Save the food macronutrient dataset as a pickle file
with open(directory_path + 'food_macronutrients_cleaned.pkl', 'wb') as f:
    pickle.dump(food_macronutrients_merged_data, f)
````
